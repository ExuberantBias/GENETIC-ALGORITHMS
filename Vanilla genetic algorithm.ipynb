```python
#Hi, this is a genetic algorithm designed to converge to a certain solution (in this case, a phrase), Â¡this gonna be fun!
#a genetic algorithm have 5 parts: 1)poblation generation 2)evaluation of each specimen (calculating their fitness according to a specific problem function)
#                                  3)selection of the best ones 4)crossover (making a new member of a new generation from 2 parents of the previous generation)
#                                  5)mutation (as the adn mutation process)
# each one of the steps from 2 to 5 are repeated N times until the problem converges to a solution previously specified
#let's start


# Initial population
objective = "im_a_phrase" #objective solution (the poblation must converge to this)

genes = len(objective) #number of genes (number of words of the objective)

import string #necesary libraries
import random 

#creates an array with the alphabet (in order to create a initial poblation first we'll need the words)
alphabet = list(string.ascii_lowercase)
alphabet.append('_')
#print(alphabet)

poblacion = [] #initializes the population


#specimen
for n in range(0,600):
    #creates the initial population
    x = ''
    for e in range(0,genes):
        i = random.randint(0,26)
        x += alphabet[i]
    
    poblacion.append(x)

print(poblacion[0:20])#shows first 20 words
```

```python
#fitness function
def fitness(x):
    #calculates the fitness of each specimen according to the correct number of letters
    k = 0
    for e in range(len(x)):
        
        if(x[e]==objective[e]):
            k+=1
  
    return k**2 
  
#fitness of each specimen
def scores(poblacion): #gives a score for each specimen according to the fitness, the index of the score in scores array corresponds to the index in population array
    score = []
    for val in poblacion: 
        score.append(fitness(val))
    for e in range(0,len(score)):
        score[e] = round((score[e]/genes)*100) #the score value now is the number of times that the specimen will appear in the mating pool
  
    return score
```

```python
def pool(scor,poblacion):
    #creates the mating pool, the value of a specimen in the score array is the number of times it will appear in the mating pool
#                        more times in mating pool, more probability of being choose as a parent
    mating_pool=[]
    for e in range(0,len(poblacion)):
        n = poblacion[e]
        for i in range(0,scor[e]):
            mating_pool.append(n)
    
    return mating_pool

#sons function ---> create a new specimen
def sons(mating_pool):
    a = random.randint(0,len(mating_pool)-1)      #chooses 2 random numbers (indexes)
    b = random.randint(0,len(mating_pool)-1)
    if (a==b):               #if the parents are the same picks a new one (could work if a and b are the same parent, but it would cause a space-time paradox)
        b = random.randint(0,len(mating_pool)-1)
    
    a = mating_pool[a]  #picks parent a form mating pool
    b = mating_pool[b] #picks parent b from mating pool
    new = crossover(a,b,len(a)) #generate a new member of the new generation
    return new
  
```

```python
def crossover(a,b,genes):#create a new specimen, a son wich genes will be replaced according with his parents genes
    new_specimen = ''
    for e in range(0,genes): #travels the fathers and pick a half of a and another half of the b
        if(e<round(genes/2)):
            new_specimen += a[e]
        else:
            new_specimen += b[e]
      
  #mutation  
    new_specimen = mutation(new_specimen) #uses mutation function ---> change a gene for a random one (0.1%)
    return new_specimen 

def mutation(x): #mutates a gene of the specimen (0.1% of probability)
    b=''
    for e in range(0,genes):
        a = random.randint(1,100)
        if (a<=10):
            i= random.randint(0,26)
            b += alphabet[i]
        else:
            b +=x[e]
  
    return b
```

```python
max_score = '' 

score = scores(poblacion)
n = 1

while(max_score!=objective):      #iterates until the best one specimen converges to the solution
    mating_pool = pool(score,poblacion)
    for e in range(0,len(poblacion)): #creates a new generation with new specimens sons of  the last one
        poblacion[e]= sons(mating_pool)
  
    score = scores(poblacion) #gives to the new specimens a score
    max_score = poblacion[score.index(max(score))] #pick the best specimen and shows it
    print('the best specimen is: ', max_score)
  
    print('fitness: ',scores([max_score]))

    n+=1
    print('generations: ',n) #show the last generation
    if(n>5000): #a limit of cycles (used to prevent an infinite loop)
        break;
  
```
